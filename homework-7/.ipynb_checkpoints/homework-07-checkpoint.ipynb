Постройте разделящие поверхности (для этого воспользуйтесь функцией `plot_surface`, пример ниже). Посчитайте accuracy на обучающей и тестовой выборках. Сильно ли деревья переобучились?

```python
def plot_surface(clf, X, y):
    plot_step = 0.01
    palette = sns.color_palette(n_colors=len(np.unique(y)))
    cmap = ListedColormap(palette)
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                         np.arange(y_min, y_max, plot_step))
    plt.tight_layout(h_pad=0.5, w_pad=0.5, pad=2.5)

    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    cs = plt.contourf(xx, yy, Z, cmap=cmap, alpha=0.3)
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap, alpha=.7,
                edgecolors=np.array(palette)[y], linewidths=2)
```

для dataset 1

как зависит алгоритим от max_depth, min_samples_leaf

```python
[X,y] = datasets[0]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
score = []
score2 = []
max_depth = [i+1 for i in range(200)]
for i in range(len(max_depth)):
    clf = DecisionTreeClassifier(max_depth = i+1)
    clf = clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    score.append(accuracy_score(y_test, y_pred))
min_samples_leaf = [i+1 for i in range(40)]
for i in range(len(min_samples_leaf )):
    clf = DecisionTreeClassifier(min_samples_leaf = i+1)
    clf = clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    score2.append(accuracy_score(y_test, y_pred))
fig, axs = plt.subplots(1,2,figsize=(10,5))
axs[0].plot(max_depth, score)
axs[0].set(xlabel = 'max_depth')
axs[0].set(ylabel = 'score')
axs[1].plot(min_samples_leaf, score2)
axs[1].set(xlabel = 'min_samples_leaf')
axs[1].set(ylabel = 'score')
fig.tight_layout(pad=5.0)
```

```python
def gini_impurity(y):
  '''
  Given a Pandas Series, it calculates the Gini Impurity. 
  y: variable with which calculate Gini Impurity.
  '''
  if isinstance(y, pd.Series):
    p = y.value_counts()/y.shape[0]
    gini = 1-np.sum(p**2)
    return(gini)

  else:
    raise('Object must be a Pandas Series.')

gini_impurity(data.Gender) 
```

```python
def entropy(y):
  '''
  Given a Pandas Series, it calculates the entropy. 
  y: variable with which calculate entropy.
  '''
  if isinstance(y, pd.Series):
    a = y.value_counts()/y.shape[0]
    entropy = np.sum(-a*np.log2(a+1e-9))
    return(entropy)

  else:
    raise('Object must be a Pandas Series.')

entropy(data.Gender)  
```
